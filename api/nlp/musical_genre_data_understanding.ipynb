```python
import seaborn as sns
import pandas as pd
import numpy as np


from model.feature_engineering import FeatureEngineering
from model.preprocessing import Preprocessing
from model.visualization import Visualization
from model.model import Model
from model.metrics import Metrics
```

```python
feature = FeatureEngineering()
preproc = Preprocessing()
visualization = Visualization()
model = Model()
metrics = Metrics()
```

```python
df_funk = pd.read_csv("dataset/funk.csv")
df_gospel = pd.read_csv("dataset/gospel.csv")
df_sertanejo = pd.read_csv("dataset/sertanejo.csv")
df_bossa_nova = pd.read_csv("dataset/bossa_nova.csv")
```

```python
df_funk_train, df_funk_valid = feature.split_train_test(df_funk, 0.2)
df_gospel_train, df_gospel_valid = feature.split_train_test(df_gospel, 0.2)
df_sertanejo_train, df_sertanejo_valid = feature.split_train_test(df_sertanejo, 0.2)
df_bossa_nova_train, df_bossa_nova_valid = feature.split_train_test(df_bossa_nova, 0.2)

df_funk_valid['genre'] = ["funk"]*len(df_funk_valid)
df_gospel_valid['genre'] = ["gospel"]*len(df_gospel_valid)
df_sertanejo_valid['genre'] = ["sertanejo"]*len(df_sertanejo_valid)
df_bossa_nova_valid['genre'] = ["bossa_nova"]*len(df_bossa_nova_valid)

df_genre_musics_valid = pd.concat([df_funk_valid, df_gospel_valid, df_sertanejo_valid, df_bossa_nova_valid])
df_genre_musics_valid.to_csv("valid/valid.csv")
```

```python
df_funk_train['lyric'] = preproc.clean_lyric(df_funk_train.lyric)
df_funk_train['famous_word'] = [feature.famous_word(lyric) for lyric in df_funk_train.lyric.to_list()]
df_funk_train['total_words'] = [feature.total_words(lyric) for lyric in df_funk_train.lyric.to_list()]
df_funk_train['percent_originality'] = [feature.percent_originality(lyric) for lyric in df_funk_train.lyric.to_list()]
df_funk_train['genre'] = ["funk"]*len(df_funk_train)

df_gospel_train['lyric'] = preproc.clean_lyric(df_gospel_train.lyric)
df_gospel_train['famous_word'] = [feature.famous_word(lyric) for lyric in df_gospel_train.lyric.to_list()]
df_gospel_train['total_words'] = [feature.total_words(lyric) for lyric in df_gospel_train.lyric.to_list()]
df_gospel_train['percent_originality'] = [feature.percent_originality(lyric) for lyric in df_gospel_train.lyric.to_list()]
df_gospel_train['genre'] = ["gospel"]*len(df_gospel_train)

df_sertanejo_train['lyric'] = preproc.clean_lyric(df_sertanejo_train.lyric)
df_sertanejo_train['famous_word'] = [feature.famous_word(lyric) for lyric in df_sertanejo_train.lyric.to_list()]
df_sertanejo_train['total_words'] = [feature.total_words(lyric) for lyric in df_sertanejo_train.lyric.to_list()]
df_sertanejo_train['percent_originality'] = [feature.percent_originality(lyric) for lyric in df_sertanejo_train.lyric.to_list()]
df_sertanejo_train['genre'] = ["sertanejo"]*len(df_sertanejo_train)

df_bossa_nova_train['lyric'] = preproc.clean_lyric(df_bossa_nova_train.lyric)
df_bossa_nova_train['famous_word'] = [feature.famous_word(lyric) for lyric in df_bossa_nova_train.lyric.to_list()]
df_bossa_nova_train['total_words'] = [feature.total_words(lyric) for lyric in df_bossa_nova_train.lyric.to_list()]
df_bossa_nova_train['percent_originality'] = [feature.percent_originality(lyric) for lyric in df_bossa_nova_train.lyric.to_list()]
df_bossa_nova_train['genre'] = ["bossa_nova"]*len(df_bossa_nova_train)
```

```python
df_genre_musics_train = pd.concat([df_funk_train, df_gospel_train, df_sertanejo_train, df_bossa_nova_train], ignore_index = True)
df_genre_musics_train.reindex()
df_genre_musics_train.to_csv("train/train.csv")
```
